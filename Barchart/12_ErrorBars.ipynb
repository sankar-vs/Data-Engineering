```python
#### Write a Python program to create bar plots with error bars on the same figure. Sample Date
#### Mean velocity: 0.2474, 0.1235, 0.1737, 0.1824
#### Standard deviation of velocity: 0.3314, 0.2278, 0.2836, 0.2645
```

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
mean_velocity = [0.2474, 0.1235, 0.1737, 0.1824]
std = [0.3314, 0.2278, 0.2836, 0.2645]
group = len(mean_velocity)
index = np.arange(group)

plt.bar(index, mean_velocity,yerr=std, color='r')
plt.ylabel('Scores')
plt.xlabel('Velocity')
plt.title('Scores by Velocity')

plt.xlim(-0.5,4)
plt.ylim(0,0.8)
plt.minorticks_on()
plt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.show()
```

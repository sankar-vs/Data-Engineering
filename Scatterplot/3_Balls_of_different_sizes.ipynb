```python
#### Write a Python program to draw a scatter plot using random distributions to generate balls of different sizes.
```

```python
from matplotlib import pyplot as plt
import random
```

```python
x = [random.random() for i in range(20)]
y = [random.random() for i in range(20)]
areas = [random.random()*100 for i in range(20)]
plt.scatter(x,y, s=areas,edgecolor='r')
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

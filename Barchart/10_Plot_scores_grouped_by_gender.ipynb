```python
#### Write a Python program to create bar plot of scores by group and gender. Use multiple X values on the same chart for men and women.
#### Sample Data:
#### Means (men) = (22, 30, 35, 35, 26)
#### Means (women) = (25, 32, 30, 35, 29)
```

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
men_means = (22, 30, 35, 35, 26)
women_means = (25, 32, 30, 35, 29)
groups = len(men_means)

fig, ax = plt.subplots()
index = np.arange(groups)
bar_width = 0.35
opacity = 0.8

barPlot1 = plt.bar(index, men_means, bar_width,alpha=opacity,color='r',label='Men')
barPlot2 = plt.bar(index + bar_width, women_means, bar_width,alpha=opacity,color='y',label='Women')

plt.xlabel('Person')
plt.ylabel('Scores')
plt.title('Scores by person')
plt.legend()

plt.show()
```

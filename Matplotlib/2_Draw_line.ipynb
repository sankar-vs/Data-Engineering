```python
'''
Write a Python program to draw a line using given axis values with suitable label in the x axis
, y axis and a title
'''
```

```python
from matplotlib import pyplot as plt
x = [1,2,3,4,5]
y = [2,8,5,6,4]
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Line Graph')
plt.show()
```

```python
##### Write a program to draw a scatter plot for random 1000 x and y coordinates
```

```python
import plotly.express as px
import numpy as np
import random
x = np.arange(1000)
y = []
for i in range(0,1000):
    n = random.randint(1,1000)
    y.append(n)
px.scatter(x=x, y=y, title="Random Plot of 1000 x and y coordinates",labels={"x": "X - axis","y":"Y - axis"})
```

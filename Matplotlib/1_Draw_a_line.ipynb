```python
'''
Write a Python program to draw a line with suitable label in the x axis, y axis and a title
'''
```

```python
from matplotlib import pyplot as plt
x = range(0,20)
y = [i*3 for i in x]
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Draw Line')
plt.show()
```
